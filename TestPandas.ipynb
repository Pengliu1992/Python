```python
import matplotlib.pyplot as plt

df=pd.read_csv('pokemon_data.csv')

L1=df['Attack']
L2=df['HP']

m1=['black' if y<50 and z<50 else 'blue' if x>0 else 'red' for x,y,z in zip((L1[L1>L2]-L2[L1>L2]),L1[L1>L2],L2[L1>L2])]

m2=['black' if y<50 and z<50 else 'blue' if x>0 else 'red' for x,y,z in zip((L1[L1<=L2]-L2[L1<=L2]),L1[L1<=L2],L2[L1<=L2])]


plt.scatter(L1[L1>L2],L2[L1>L2],c=m1,s=abs(L1[L1>L2]-L2[L1>L2]),marker='^',label='Attack Type')

plt.scatter(L1[L1<=L2],L2[L1<=L2],c=m2,s=abs(L1[L1<=L2]-L2[L1<=L2]),marker='o',label='Tank Type')

plt.legend()
```
